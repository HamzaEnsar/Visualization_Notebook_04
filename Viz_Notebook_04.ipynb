**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python

sns.boxplot(x='gender', y='age', hue='manner_of_death', data=kill)
plt.title('Quartiles of Age by Gender and Manner of Death')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()


```

```python


sns.swarmplot(x='gender',y='age',hue='manner_of_death',data=kill)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python

state_freq = kill.state.value_counts()[:20]
state_freq

fig, ax = plt.subplots(figsize=(8, 5))

ax = sns.barplot(x=state_freq.index, y=state_freq.values);
```

```python




fig, ax = plt.subplots(figsize=(8, 5))

sns.countplot(x="state", 
              data=kill,
              order=state_freq.index)

for i, count in enumerate(state_freq):
    plt.text(i, count, str(count), ha='center', va='bottom');
```
